```python
def print_directory_contents(sPath):
    import os
    for child in os.listdir(sPath):
        child_path = os.path.join(sPath,child)
        if os.path.isdir(child_path):
            print_directory_contents(child_path)
        else:
            print(child_path)
```

```python
print_directory_contents("./")
```
